```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
```

```python
def plot(exp_name, x, y):
    filenames = [f"../out/{exp_name}_{i+1}.pkl" for i in range(2)]
    dfs = [pd.read_pickle(f) for f in filenames]
    df = pd.concat(dfs)
    plt.figure(figsize=(10, 5))
    sns.lineplot(x=x, y=y, hue="method", data=df)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plot("deletion_curves_mnist", "mask", "difference")
```

```python
plot("insertion_curves_mnist", "insertion", "difference")
```
