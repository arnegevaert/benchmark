```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
from os import path
import os
mpl.rcParams['figure.dpi'] = 150
```

```python
def lineplot(filenames, x, y):
    dfs = [pd.read_pickle(f) for f in filenames]
    df = pd.concat(dfs)
    plt.figure(figsize=(7, 4))
    sns.lineplot(x=x, y=y, hue="method", data=df)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
def boxplot(filenames, x, y, xtick_rotation):
    dfs = [pd.read_pickle(f) for f in filenames]
    df = pd.concat(dfs)
    plt.figure(figsize=(7, 4))
    ax = sns.boxplot(x=x, y=y, data=df, showfliers=False)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=xtick_rotation)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
lineplot("deletion_curves_mnist", "mask", "difference")
```

```python
fns = ["../out/mnist/insertion_curves_mnist.pkl"]
lineplot(fns, "insertion", "difference")
```

```python
boxplot("infidelity_mnist", "method", "infidelity", xtick_rotation=30)
```
