```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
```

```python
def lineplot(exp_name, x, y):
    filenames = [f"../out/{exp_name}_{i+1}.pkl" for i in range(2)]
    dfs = [pd.read_pickle(f) for f in filenames]
    df = pd.concat(dfs)
    plt.figure(figsize=(10, 5))
    sns.lineplot(x=x, y=y, hue="method", data=df)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
def boxplot(exp_name, x, y, xtick_rotation):
    filenames = [f"../out/{exp_name}_{i+1}.pkl" for i in range(2)]
    dfs = [pd.read_pickle(f) for f in filenames]
    df = pd.concat(dfs)
    plt.figure(figsize=(10, 5))
    ax = sns.boxplot(x=x, y=y, data=df, showfliers=False)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=xtick_rotation)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
lineplot("deletion_curves_mnist", "mask", "difference")
```

```python
lineplot("insertion_curves_mnist", "insertion", "difference")
```

```python
boxplot("infidelity_mnist", "method", "infidelity", xtick_rotation=30)
```

```python
lineplot("sensitivity_n_mnist", "n", "correlation")
```
