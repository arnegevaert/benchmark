```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logitsoftmax(x):
    return x - np.log(np.sum(np.exp(x), axis=-1, keepdims=True) - np.exp(x))
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x_range = np.linspace(-10, 10, 500)
X = np.ones((500, 5))
X[:, 3] = x_range
X[:, 4] = 2*x_range
lgsm = logitsoftmax(X)
```

```python
fig, ax = plt.subplots()
ax.grid()
ax.set_xticks(np.arange(-10, 10, 1.))
ax.set_yticks(np.arange(int(np.min(lgsm)), int(np.max(lgsm)), 1.))
ax.plot(x_range, lgsm[:, 3])
```

```python
x_range_part = x_range[np.where((x_range < 1.5) & (x_range > 0.5))]
lgsm_part = lgsm[np.where((x_range < 1.5) & (x_range > 0.5))]
fig, ax = plt.subplots()
ax.grid()
ax.plot(x_range_part, lgsm_part[:, 3])
```
